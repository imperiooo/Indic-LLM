```python
def correctRating(x):
    if len(re.findall(r".+/\s*5", str(x))):
        s = re.sub(r"/\s*5", '', re.findall(r".+/\s*5", str(x))[0].replace('रेटिंग','').replace(':','').strip())
        if len(s) > 3:
            s = s[-3:]
        return s
    else:
        return ''
```

```python
df['rating'] = df['rating'].apply(correctRating)
```

```python
# Binary Classification problem -- if rating <= 1.5 => Negative review (0), if rating >=3 Positive review(2), else Neutral review(1)
```

```python
df = pd.read_csv('hindi_movie_review_cleaned.csv', lineterminator='\n')
```

```python
df.head()
```

```python
# Initializing everyone's experience with 1 (i.e Neutral)
df['experience'] = pd.Series([1]*df.shape[0], index=df.index)
```

```python
# Those with rating <= 1.5, experience is 0 (i.e Bad)
df.loc[df['rating'] <= 1.5, 'experience'] = 0
```

```python
# Those with rating >= 3, experience is 2 (i.e Good)
df.loc[df['rating'] >= 3, 'experience'] = 2
```

```python
df_final = {'text': df['article'], 'experience': df['experience']}
```

```python
df_final = pd.DataFrame(df_final)
```

```python
df_final.head()
```

```python
df_final.to_csv('Hindi_Classification_Data.csv', index=False)
```
