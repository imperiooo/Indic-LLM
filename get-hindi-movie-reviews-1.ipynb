```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import re
import pickle
number_of_pages = 38
all_articles = []
```

```python
import urllib.parse
```

```python
for i in range(1, number_of_pages+1):
    url = 'http://hindi.webdunia.com/bollywood-movie-review/' + str(i)
    html_doc = ''
    with urlopen(url) as response:
        for line in response:
            line = line.decode('utf-8')
            html_doc = html_doc + line.replace('\n','')
    soup = BeautifulSoup(html_doc, 'html.parser')
    base_url = 'http://hindi.webdunia.com'
    first_article = soup.find('div',{'class':'list list_l wbx rel'})['onclick']
    first_article = first_article[13:-12]
    all_articles.append(base_url + first_article)
    rest_articles_div = soup.find_all('div', {'class':'col-sm-6 listItem'})
    for div in rest_articles_div:
        all_articles.append(base_url + div.find('a')['href'])
    print('For ' + url + ' ' + str(len(all_articles)))
```

```python
# We have ratings from -17
articlesAndRating = {}
for c, url in enumerate(all_articles):
    html_doc = ''
    url = urllib.parse.urlsplit(url)
    url = list(url)
    url[2] = urllib.parse.quote(url[2])
    url = urllib.parse.urlunsplit(url)
    with urlopen(url) as response:
        for line in response:
            line = line.decode('utf-8')
            html_doc = html_doc + line.replace('\n','')
    soup = BeautifulSoup(html_doc, 'html.parser')
    main_div = soup.find('div',{'id':'ram'})
    divs = main_div.find_all('div', {'class':None, 'id':None})
    divs = divs + soup.find_all('font')
    if len(divs) == 0:
        divs = soup.find_all('div')
    article = ''
    rating = ''
    for div in divs:
        if 'google' in div.text or ('रेटिंग' and '/5') in div.text:
            if ('रेटिंग' and '/5') or ('रेटिंग' and '/ 5') in div.text:
                rating = div.text.strip(' \t\n\r').replace(u'\xa0', u' ').replace(u'\ufeff','')
            continue
        article = article + div.text.strip(' \t\n\r') + '\n'
    if rating is '':
        elems = soup(text=re.compile(r'रेटिंग'))
        if len(elems):
            rating = elems[0]
        else:
            # calling again with second url
            url = url[:-5] + str(2) + url[-4:]
            html_doc = ''
            url = urllib.parse.urlsplit(url)
            url = list(url)
            url[2] = urllib.parse.quote(url[2])
            url = urllib.parse.urlunsplit(url)
            try:
                with urlopen(url) as response:
                    for line in response:
                        line = line.decode('utf-8')
                        html_doc = html_doc + line.replace('\n','')
            except:
                continue
            soup = BeautifulSoup(html_doc, 'html.parser')
            main_div = soup.find('div',{'id':'ram'})
            divs = main_div.find_all('div', {'class':None, 'id':None})
            divs = divs + soup.find_all('font')
            if len(divs) == 0:
                divs = soup.find_all('div')
            rating = ''
            for div in divs:
                if 'google' in div.text or ('रेटिंग' and '/5') in div.text:
                    if ('रेटिंग' and '/5') or ('रेटिंग' and '/ 5') in div.text:
                        rating = div.text.strip(' \t\n\r').replace(u'\xa0', u' ').replace(u'\ufeff','')
                    continue
                article = article + div.text.strip(' \t\n\r') + '\n'
            if rating is '':
                elems = soup(text=re.compile(r'रेटिंग'))
                if len(elems):
                    rating = elems[0]
    
    article = article.strip(u'\u200b')
    article = re.sub(r'^https?:\/\/.*[\r\n]*', '', article)
    article = article.replace(u'\ufeff','')
    article = article.replace(u'\xa0', u' ')
    article = article.replace('  ', ' ');
    article = article.replace(' , ', ', ');
    articlesAndRating[c] = {
        'url': url,
        'article': article,
        'rating': rating
    }
    print('Saved ' + str(c) + ' with rating ' + rating)
```
